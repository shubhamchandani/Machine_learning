```python
from bs4 import BeautifulSoup as bs
from urllib import request
import re
import nltk
print(1)
```

```python
web_data=request.urlopen('https://webkul.com/')
#web_data.read() will  give all the source data
```

```python
soup=bs(web_data,'html5lib')
```

```python
data = soup.findAll(['h1','h3','h2','h4','h5','h6','p'])
```

```python
for i in range(len(data)):
    data[i]=data[i].text
```

```python
data = ' '.join(data)
```

```python
#  dATA cleaning
clean_data= re.sub(r'\[[0-9]*\]',' ',data)  #  this will remove 0 or more times number appearing in mydata
clean_data= re.sub(r'\s+',' ',clean_data)  # it will remove one or more white spaces with single white space
clean_data= re.sub(r'[^a-zA-Z]',' ',clean_data)  # it will remove single char from starting of the line
clean_data= re.sub(r'\s+',' ',clean_data)  # it will remove one or more white spaces with single white space
```

```python
clean_data=clean_data.split(" ")
```

```python
from nltk.corpus import stopwords
l=[]
for i in clean_data:
    if(i.lower() not in stopwords.words('english')):
        l.append(i)
```

```python
a=nltk.FreqDist(l)
```

```python
a.plot(30)
```
